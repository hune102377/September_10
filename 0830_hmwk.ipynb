#### 1. k-NN 알고리즘의 neighbor 개수에 따른 socre 값 변화 확인
- 1장에서 n_neighbors의 값을 5부터 49까지 바꾸어 가면서 score() 함수의 결과 및 n_neighbors의 값을 plot 차트로 표현하시오

```python
# 기본적인 데이터 설정\
# 도미 length Data
bream_length =	[25.4,	26.3,	26.5,	29.0,	29.0,	29.7,	29.7,	30.0,	30.0,	30.7,	31.0,	
31.0,	31.5,	32.0,	32.0,	32.0,	33.0,	33.0,	33.5,	33.5,	34.0,	34.0,	
34.5,	35.0,	35.0,	35.0,	35.0,	36.0,	36.0,	37.0,	38.5,	38.5,	39.5,	
41.0,	41.0]

# 도미 weight Data
bream_weight = [242.0,	290.0,	340.0,	363.0,	430.0,	450.0,	500.0,	390.0,	450.0,	500.0,	
475.0,	500.0,	500.0,	340.0,	600.0,	600.0,	700.0,	700.0,	610.0,	650.0,
575.0,	685.0,	620.0,	680.0,	700.0,	725.0,	720.0,	714.0,	850.0,	1000.0,
920.0,	955.0,	925.0,	975.0,	950.0]
```

```python
# 빙어 length 데이터
smelt_length = [9.8, 10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2,
12.4, 13.0, 14.3, 15.0]

# 빙어 weight 데이터
smelt_weight = [6.7, 7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4,
12.2, 19.7, 19.9]
```

```python
# 길이 데이터와 무게 데이터 합치기
length = bream_length + smelt_length
weight = bream_weight + smelt_weight
```

```python
# 1차원 리스트 2개를 2차원 리스트로 변경
# zip(length, weight) 함수로 이용
fish_data = [[l, w] for l, w in zip(length, weight)]
```

```python
# 정답 데이터(fish_target) 준비
# 생선의 길이와 무게를 이용하여 도미(1)와 빙어(0)를 구분
fish_target = [1] * 35 + [0] * 14
```

```python
# 수식에 사용될 변수 준비
list_num = list(range(5, 50))
print(list_num)
```

```python
from sklearn.neighbors import KNeighborsClassifier

# score 점수를 담을 리스트 생성
list_score = []

# list_num 변수별로 score() 점수 계산 후 저장
for i in range(0, len(list_num)):
    kn49 = KNeighborsClassifier(n_neighbors=list_num[i])
    kn49.fit(fish_data, fish_target)
    list_score.append(kn49.score(fish_data, fish_target))
```

```python
# 시각화 및 플래폼 라이브러리 import
import matplotlib.pyplot as plt
import	platform

# 시스템별 한글 폰트 설정
if	platform.system()	==	'Windows':
    plt.rc('font', family='Malgun Gothic')
elif platform.system()	==	'Darwin':	#	MacOS
    plt.rc('font', family='AppleGothic')
else:
    plt.rc('font', family='AppleGothic')

# 선 그래프 그리기
plt.plot(list_num, list_score, color='blue')
# 산점도 그래프 그리기
plt.scatter(list_num, list_score, color='blue')
# 라벨 설정
plt.xlabel('neighbor')
plt.ylabel('score')
# 타이틀 설정
plt.title('neighbor - score')
# 표시
plt.show()
```

#### 2. plot()에 사용한 n_neighbors와 score 값을 이용하여 n_neighbors의 개수에 대한 score 값을 numpy의 column_stack() 함수를 이용하여 ndarry 타입으로 변환한 다음, 아래와 같이 pandas의 DataFrame으로 변환하여 출력하시오

```python
# numpy 라이브러리 import
import numpy as np

# column_stack 을 활용하여 두 데이터 리스트를 total_data라는 numpy.ndarray로 만듬
total_data = np.column_stack((list_num,list_score))
```

```python
# pandas 라이브러리 import
import pandas as pd

# total_data 데이터를 데이터 프레임으로 변환
df = pd.DataFrame(total_data, columns = ['neighbors', 'score'])

df
```
