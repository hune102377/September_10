```python
# 주소 예시
example_url_01 = 'https://n.news.naver.com/article/088/0000777435?cds=news_media_pc&type=editn'
example_url_02 = 'https://n.news.naver.com/article/437/0000315890?cds=news_media_pc&type=editn'
example_url_03 = 'https://n.news.naver.com/article/448/0000376369?cds=news_media_pc&type=editn'
example_url_04 = 'https://n.news.naver.com/article/088/0000777295?cds=news_media_pc&type=editn'
example_url_05 = 'https://n.news.naver.com/article/657/0000007245?cds=news_media_pc&type=editn'
```

```python
# 모듈 불러오기
from urllib.request import urlopen #url 주소 호출 라이브러리
from bs4 import BeautifulSoup
import numpy as np
```

```python
html = urlopen(example_url_01) # url 주소 html로 저장
soup = BeautifulSoup(html.read(), 'html.parser') # html 데이터 BeautifulSoup으로 요약
a = soup.contents[2].text
t = a.split('\n')
print(t[4])
```

```python
def getNews(url) :
    html = urlopen(url) # url 주소 html로 저장
    soup = BeautifulSoup(html.read(), 'html.parser') # html 데이터 BeautifulSoup으로 요약
    a = soup.contents[2].text
    t = a.split('\n')

    list_len = []
    for i in range(0, len(t)):
        list_len.append(len(t[i]))
    num = np.argmax(list_len)

    title = soup.title.text

    print(title)
    print(t[num])
```

```python
getNews(example_url_01)
```

```python
getNews(example_url_02)
```

```python
getNews(example_url_03)
```
