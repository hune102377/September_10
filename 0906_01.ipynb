## 보스턴 집값 가지고 놀기

```python
import pandas as pd
import numpy as np

data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
target = raw_df.values[1::2, 2]
```

```python
data.shape, target.shape
```

### 우선은 logistic 한번 해보기

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR

train_input, test_input, train_target, test_target = train_test_split(
    data, target, random_state= 42
)

modelSVR = SVR().fit(train_input, train_target)


print("훈련 점수 :", modelSVR.score(train_input, train_target))
print("테스트 점수 :", modelSVR.score(test_input, test_target))
```

```python
from sklearn.linear_model import SGDRegressor
modelSGD = SGDRegressor().fit(train_input, train_target)

print("훈련 점수 :", round(modelSGD.score(train_input, train_target), 3))
print("테스트 점수 :", round(modelSGD.score(test_input, test_target), 3))
```

점수가 너무 낮다 다른 모델을 찾아보자

```python
from sklearn.utils import all_estimators
import warnings
warnings.filterwarnings('ignore')

all_algorithms = all_estimators(type_filter='regressor')


list_name_model = []
list_train_score = []
list_test_score = []


for (name, algorithm) in all_algorithms:
    try :
        clf = algorithm()
        clf.fit(train_input, train_target)

        list_train_score.append(round(clf.score(train_input,train_target), 3))
        list_test_score.append(round(clf.score(test_input,test_target)))
        list_name_model.append(name)
    except :
        continue
```

```python
df = pd.DataFrame([list_name_model, list_train_score, list_test_score]).T
df.columns = ['모델', '훈련 점수', '테스트 점수']
```
